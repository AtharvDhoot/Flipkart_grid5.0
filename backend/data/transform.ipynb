```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./articles.csv")
data.head()
```

```python
data.columns
```

```python
columnsToDrop = []

for col in data.columns:
    if col.split('_')[-1] in ["id", "code", "no"] and col not in ['article_id', 'product_code']:
        columnsToDrop.append(col)

data = data.drop(columnsToDrop, axis=1)
data.columns
```

```python
import json

# Open the JSON file for reading
with open('product_score_map.json', 'r') as json_file:
    scoreMap = json.load(json_file)
```

```python
trendScores = []

for code in data.product_code:
    productId = f'0{code}'

    if scoreMap.get(productId) is not None:
        trendScores.append(scoreMap[productId])
    else:
        trendScores.append(0)

trendScores
```

```python
data["trend_score"] = trendScores

data.head()
```

```python
data.to_csv("./products.csv", index=False)
```
